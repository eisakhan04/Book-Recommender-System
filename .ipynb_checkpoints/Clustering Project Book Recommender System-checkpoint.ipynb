```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index==book_name)[0][0]
    distances , suggestions=model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)
    
    for i in range(len(suggestions)):
        if i==0:
            print("The Suggestions for " , book_name , "are :")
        if not i:
            print(book_pivot.index[suggestions[i]])
```

```python
recommend_book("Harry Potter and the Chamber of Secrets (Book 2)")
```

# Popularity Based Recommender System 

```python
ratings.shape
```

```python
book.head(2)
```

```python
ratings.head(2)
```

```python
ratings_with_name= ratings.merge(book , on='ISBN')
```

```python
ratings_with_name.shape
```

```python
ratings_with_name.head(2)
```

```python
ratings_with_name.groupby('title').count()
```

```python
num_rating_df = ratings_with_name.groupby('title').count()['rating'].reset_index()
num_rating_df.rename(columns={'rating':'num_rating'} , inplace=True)
```

```python
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('title').mean()['rating'].reset_index()
avg_rating_df.rename(columns={'rating':'avg_rating'} , inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_rating']>=100].sort_values('avg_rating' , ascending=False).head(50)
```

```python
book.head()
```

```python
popular_df = popular_df.merge(book , on='title').drop_duplicates('title')[['title' , 'author' , 'Image_URL_M' , 'num_rating' , 'avg_rating']]
```
